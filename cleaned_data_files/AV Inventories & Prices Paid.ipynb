Load the packages and create the dataframe from the CSV in the Resouces folder


```python
import pandas as pd
ico_inventories = "../Resources/inventories.csv"
ico_price_to_grower = "../Resources/prices-paid-to-growers.csv"
ico_inventories_df = pd.read_csv(ico_inventories)
ico_price_to_grower_df = pd.read_csv(ico_price_to_grower)
```

```python
ico_inventories_df
```

```python
ico_price_to_grower_df
```

```python
ico_inventories_df=ico_inventories_df.rename(columns={"inventories": "Country"})
ico_price_to_grower_df=ico_price_to_grower_df.rename(columns={"prices_paid_to_growers": "Country"})
ico_inventories_df=ico_inventories_df.dropna()
ico_price_to_grower_df=ico_price_to_grower_df.dropna()
ico_inventories_ten=ico_inventories_df[["Country", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018"]]
ico_price_to_grower_ten=ico_price_to_grower_df[["Country", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018"]]

ico_price_to_grower_ten = ico_price_to_grower_ten.transpose()
ico_price_to_grower_ten = ico_price_to_grower_ten.reset_index(drop=False)
new_header = ico_price_to_grower_ten.iloc[0]
ico_price_to_grower_ten = ico_price_to_grower_ten[1:]
ico_price_to_grower_ten.columns = new_header 
ico_price_to_grower_ten=ico_price_to_grower_ten.rename(columns={"Country": "Year"})
ico_price_to_grower_ten=ico_price_to_grower_ten.loc[:, (ico_price_to_grower_ten != 0).any(axis=0)]

ico_inventories_ten = ico_inventories_ten.transpose()
ico_inventories_ten = ico_inventories_ten.reset_index(drop=False)
new_header = ico_inventories_ten.iloc[0]
ico_inventories_ten = ico_inventories_ten[1:]
ico_inventories_ten.columns = new_header 
ico_inventories_ten=ico_inventories_ten.rename(columns={"Country": "Year"})
ico_inventories_ten=ico_inventories_ten.loc[:, (ico_inventories_ten != 0).any(axis=0)]
```
